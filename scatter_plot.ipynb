```python
import matplotlib.pyplot as plt

# Sample data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
colors=['#f7e40c' if score>50 else '#fa1007' for score in exam_scores]
size=[score*2 for score in exam_scores]
plt.scatter(study_hours, exam_scores,c=colors,s=size,)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
# using color map in scatter plot 


import matplotlib.pyplot as plt

# Sample data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 99, 50, 55, 60, 65, 75, 85, 90]

social_cores=[10,500,200,300,567,999,454,742,600]
plt.scatter(study_hours, exam_scores,c=social_cores,cmap='plasma')

# color mapping is according to the social_scores not exam scores
plt.colorbar(label='social score')
for i in range(len(study_hours)):   
    plt.text(study_hours[i]-0.2,exam_scores[i]-2.3,f'std{i+1}')# use + or - for adjusting the position according to x,y cordinates 
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True,linestyle='--',color='black',alpha=0.6)
plt.show()
# plt.text( x + or - position, y + or - postion , label
# plt.annotate(f'label{i+1} , (x-axis variable, y-axis variable)) 
```

```python
# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='Class A', c='blue')
plt.scatter(class_b_hours, class_b_scores, label='Class B', c='red')

plt.title('Scatter Plot: Class A vs Class B',fontsize=16)
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True,color='black',linestyle='--',alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['Aman', 'Neha', 'Ravi', 'Pooja', 'Kunal', 'Mehak', 'Tanish']
coding_hours = [2, 4, 1, 3, 5, 2, 6]
marks = [55, 70, 50, 65, 80, 60, 90]
size=[score*3 for score in marks]

plt.figure(figsize=(8, 5))
scatter = plt.scatter(coding_hours, marks, c=marks, cmap='cool', s=90, edgecolors='k')

for i in range(len(students)):
    plt.text(coding_hours[i]-0.2, marks[i]-1.9, students[i], fontsize=9)

plt.xlabel('Coding Hours per Day')
plt.ylabel('Marks (%)')
plt.title('Coding vs Marks (Scatter Plot)')
plt.colorbar(scatter, label='Marks')
plt.grid(axis='y',color='black',linewidth=1,alpha=0.6,linestyle='--')
plt.show()
plt.tight_layout()
plt.savefig('coding.png',dpi=300,bbox_inches='tight')
```

```python
age = [5, 8, 12, 15, 18, 22, 25, 28, 30, 35]
height = [7, 10, 14, 17, 20, 24, 28, 30, 33, 37]
plt.figure(figsize=(10,6))
plt.scatter( age , height ,c=height,s=120,cmap='viridis')
for i in range(len(age)):
    plt.text(age[i]+0.4,height[i]-1.1, height[i],color='#6ffc95',fontweight='bold',fontsize=10)

plt.colorbar(label='height')
plt.grid(axis='y',color='white', alpha=0.6, linestyle='--',linewidth=2)
plt.xticks(fontsize=14,color='cyan',fontweight='heavy')
plt.yticks(fontsize=14,color='cyan',fontweight='heavy')
plt.xlabel('Age',fontsize=16,color='yellow',fontweight='bold')
plt.ylabel('Height',fontsize=16,color='yellow',fontweight='bold')
plt.title('Height according to age', fontsize=18,color='orange',fontweight='black')
plt.style.use('dark_background')
plt.savefig('height_vs_age.png',dpi=300)

```
