```python
# basic pie chart example:

import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')
```

```python
# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
plt.show()
```

```python
# changing diffrent colors 
colors= ['#68f78e' , '#68cef7' , '#f56767', '#f5ba67']
plt.pie(runs, labels=labels, colors=colors)
```

```python
# adding styles and edges in the pie chart using wedgeprops
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--','fill':True}
)
plt.show()

# wedgeprops = { 'edgecolor': , 'linewidht', }
```

```python
# explode keyword use to pull out one or more slices
explode = [0.1, 0.4, 0, 0]  # highlighting sachin and sehwag slice using explode

plt.pie(runs, labels=labels, explode=explode, colors=colors, wedgeprops= { 'edgecolor': 'black', 'linestyle':'--','linewidth':2})
plt.title("Exploded Pie Example", fontweight='light',fontstyle='italic',fontsize=16,color='#a834fa')
```

```python
# adding shadows for 3 D feel
new = [ 0.1,0,0,0]
plt.pie(
    runs,
    labels=labels,
    explode=new,
    colors=colors,
    shadow=True,    # for 3D shadow feel use shadow parameter=True
    startangle=180  # to rotate your pie chart at particular angle
)
#
```

```python
# to see percentage of a pie chart by using autopct
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,
    colors=['#68f78e' , '#68cef7' , '#f56767', '#f5ba67'],
    wedgeprops={'linewidth':2, 'edgecolor':'black', 'linestyle':'--'}
)
plt.title("Career Run Share")
plt.show()
```

```python
# my first assignment 
# Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understand.

os_names = ['Android', 'iOS', 'KaiOS', 'HarmonyOS', 'Windows', 'Blackberry', 'Others']
market_shares = [71, 27, 0.6, 0.4, 0.2, 0.1, 0.7]
os_grouped = ['Android', 'iOS', 'Others']
market_grouped = [71, 27, 2.0]
plt.pie( market_grouped , labels=os_grouped , colors=['#34a4fa','#f7f374','#f0603c'] ,
        shadow=True, wedgeprops= { 'edgecolor':'black','linewidth':2, 'linestyle':'--'},autopct='%1.1f%%',explode=[0.1,0,0])
# plt.text(-1.5, -1, "Note:\n'Others' includes KaiOS, HarmonyOS, Windows, Blackberry", fontsize=9, color='gray')

plt.title( ' OS sales percentage ', fontstyle='italic',fontsize=16,fontweight='heavy',color='#565ef5')
plt.tight_layout()        
```

```python
# Horizontal bar chart
plt.barh(os_grouped,market_grouped,color=['#34a4fa', '#f7f374', '#f0603c'])
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share - Horizontal Bar Chart", fontstyle='italic', fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle=':', alpha=0.6)
plt.show()
plt.tight_layout()
# alpha = 0.6 used to maintain t
```
