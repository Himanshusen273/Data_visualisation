```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
plt.show()
# syntax = plt.plot(x-axis,y-axis)
```

```python
# marking the graphs and adding lables
plt.plot(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("score comparison")
plt.legend()
plt.show()


# bad method tyytiyitutuiyuty[tuyartyriuriur jirgir'girgu'r[
# plt.legend(["Kohli", "Sehwag"])
```

```python
plt.plot(years, kohli, 'ro--', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'go--', label="Sehwag")  # green triangles dotted
plt.legend()
plt.grid(True)
plt.xlabel('year')
plt.ylabel('runs')
plt.title('comparison graph')
plt.show()
# syntax of format string in matplotlib
# 'color,marker,line'
```

```python
# practicing format string
# Sample Data
years = [2015, 2016, 2017, 2018, 2019, 2020]
kohli = [1200, 950, 1100, 1300, 1050, 980]
dhoni = [800, 850, 780, 700, 650, 600]

# style



# Plotting
plt.plot(years, kohli,'r^--',label='Kohli',linewidth=2)
plt.plot(years, dhoni,'yD--', label='Dhoni',linewidth=2)

# Labels & Title
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs Scored Over the Years")

# Legend
plt.legend()

# Show Plot
plt.grid(True)

plt.show()
```

## plt.style.available # to check which styles are available for plotting

```python
import matplotlib.pyplot as plt
# X-axis ke liye years
years = [2015, 2016, 2017, 2018, 2019, 2020]

# Alag-alag players ke runs (dummy data)
kohli = [700, 850, 750, 900, 880, 670]
devilliers = [600, 780, 720, 810, 790, 620]
gayle = [500, 690, 640, 730, 710, 580]

plt.plot( years , kohli , 'ro--' , linewidth=2 , label='virat_runs')
plt.plot(years, devilliers, 'g>-.', linewidth=2 , label='devilliers_runs')
plt.plot( years, gayle, 'yD:' , linewidth=2, label='gayle_runs')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
plt.tight_layout()
plt.xlabel('years', fontstyle='italic',fontsize=12)
plt.ylabel('scores',fontstyle='oblique',fontsize=12)
plt.title('Score--comparison', fontsize=16, fontweight='bold')
plt.savefig('score_compare.png', dpi=300, bbox_inches='tight')
```

```python
years = [2018, 2019, 2020, 2021, 2022]

rohit = [680, 790, 720, 850, 810]
dhoni = [600, 550, 580, 610, 590]
kl_rahul = [500, 670, 690, 730, 750]
pant = [450, 490, 530, 600, 640]

plt.plot(years, rohit , 'b>:', linewidth=3, label='virat')
plt.plot(years,dhoni, 'g>:', linewidth=3, label='dhoni')
plt.plot(years,kl_rahul, 'yD:', linewidth=3, label='rahul')
plt.plot(years, pant, 'm*--', linewidth=3, label='rishabh')
plt.style.use('dark_background')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.xlabel('years',fontstyle='italic',fontsize=12,fontweight='bold',color='cyan')
plt.ylabel('Runs', fontstyle='italic',fontsize=12,fontweight='heavy',color='orange')
plt.title('career_comparison',fontsize=16, fontstyle='oblique',fontweight='black',color='red')
plt.savefig('score_chart.png',dpi=300, bbox_inches='tight')
```
