```python

# bar charts 
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.bar(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs", fontweight='light')
```

```python
import numpy as np

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))  # index positions
width = 0.25

plt.bar(x - width, sachin, width=width, label="Sachin" )
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison",fontweight='heavy',color='red')
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt

name = [ 'himanshu', 'sabarwal' , 'dhruv' , 'abhishek', 'varun']
scores = [ 183 ,145, 167, 256, 186 ]

plt.bar( name , scores , width= 0.56 , color='yellow')

for i in range(len(name)):
    plt.text( i , scores[i] - 50 , str(scores[i]), color='red' , fontsize=15 , fontweight='heavy' , ha ='center',va='bottom')

plt.tight_layout()
plt.grid(False)
plt.style.use('dark_background')

plt.title('score-comparison', color='cyan', fontsize=18 , fontstyle='italic', fontweight='black')
plt.ylabel('scores' , color='orange' , fontsize=15, fontstyle='normal' , fontweight='light')
plt.savefig( 'bar_graph.png', dpi=300 , bbox_inches='tight')
# adding labels in the baar 
```
