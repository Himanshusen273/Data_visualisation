```python
import matplotlib.pyplot as plt

activities = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
time_spent = [3, 2, 2, 5]  # hours in a day

colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']

plt.figure(figsize=(6,6))
plt.pie(time_spent, labels=activities, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("After School Activities")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python


days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['red', 'blue', 'green', 'yellow']

plt.figure(figsize=(10,6))
plt.stackplot(days, studying, playing, watching_tv, sleeping,
              labels=labels, colors=colors, alpha=0.5)

plt.legend(loc='upper left')  # Location of the legend
plt.title('Weekly Activity Tracker')
plt.xlabel('Day')
plt.ylabel('Hours')
plt.grid(color='black',linewidth=1,linestyle='-.',alpha=0.6)
plt.show()
```

```python
# practicing stackplot to understand the concepts more 


import matplotlib.pyplot as plt

# Days
days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7']

# Data for each task
sleeping = [8, 7, 6, 8, 7, 8, 9]
coding = [4, 5, 6, 5, 6, 5, 4]
workout = [1, 1, 1, 1, 1, 1, 1]
eating = [2, 2, 2, 2, 2, 2, 2]
playing = [2, 3, 3, 2, 2, 3, 2]

# Stackplot
labels = ['Sleeping', 'Coding', 'Workout', 'Eating', 'Playing']
colors = ['red', 'yellow', 'green', '#f1faee', '#ffb703']

plt.figure(figsize=(10,6))
plt.stackplot(days, sleeping, coding, workout, eating, playing, labels=labels, colors=colors, alpha=0.6)

plt.xlabel("Days", fontsize=14, fontweight='bold', fontstyle='italic')
plt.ylabel("Hours Spent", fontsize=14, fontweight='bold', fontstyle='italic')
plt.title("Weekly Task Distribution", fontsize=16, fontweight='bold', color='#1d3557')
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(linestyle='--', alpha=0.7, linewidth=1,color='black')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
