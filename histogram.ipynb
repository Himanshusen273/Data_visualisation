```python
import matplotlib.pyplot as plt
import numpy as np

ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]

bins = [ 15, 20,25,30,35,40,45]

plt.hist(ages, bins=bins, edgecolor='gold',linestyle='--',linewidth=2)
plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5,linewidth=1,color='black',axis='y')
plt.show()
```

```python
import numpy as np
workout_minutes = [30, 45, 50, 20, 60, 35, 40, 55, 25, 45, 60, 30, 30, 50, 40, 35, 55, 60, 45, 30, 50, 40, 25, 30, 35, 45, 50, 60, 40, 30]
bins = [10,20,30,40,50,60]
plt.hist( workout_minutes, bins=bins , edgecolor='red',linewidth=1,linestyle='--',color='green',alpha=0.6)
plt.axvline(np.mean(workout_minutes), color='red', linestyle='--', linewidth=2, label='Average Age')
plt.xlabel('minutes',fontsize=12)
plt.ylabel('frequency',fontsize=12)
plt.tight_layout()
plt.grid( color='black',linewidth=1,linestyle='--')
plt.xticks(rotation=45,color='red',alpha=0.8)
plt.yticks(rotation=45,color='blue',alpha=0.8)


```
